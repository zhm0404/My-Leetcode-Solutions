```python
# Given an integer array nums, return all the triplets [nums[i], nums[j], nums[k]] such that i != j, i != k, and j != k, and nums[i] + nums[j] + nums[k] == 0.
# Notice that the solution set must not contain duplicate triplets.
class Solution(object):
    def threeSum(self, nums):
        nums.sort()
        s = set()
        output = []
        for i in range(len(nums)):
            j = i + 1
            k = len(nums) - 1
            while j < k:
                if nums[i] > 0 or nums[k] < 0:
                    break;
                sum = nums[i] + nums[j] + nums[k]
                if  sum == 0:
                    s.add((nums[i], nums[j], nums[k]))
                    j += 1
                    k -= 1
                elif sum < 0:
                    j += 1
                else:
                    k -= 1
        output = list(s)
        return output
# class Solution(object):
#     def threeSum(self, nums):
#         """
#         :type nums: List[int]
#         :rtype: List[List[int]]
#         """
#         result = []
#         nums = sorted(nums)
#         for i in range(len(nums)):
#             numToIndex = {}
#             for j in range(i+1, len(nums)):
#                 if 0 - nums[i] - nums[j] in numToIndex:
#                     result.append(tuple([nums[i], nums[j], 0 - nums[i] - nums[j]]))
#                 numToIndex[nums[j]] = j
#         return set(result)
# class Solution:
#     def threeSum(self, nums):
#         target = 0
#         nums.sort()
#         s = set()
#         output = []
#         for i in range(len(nums)):
#             j = i + 1
#             k = len(nums) - 1
#             while j < k:
#                 sum = nums[i] + nums[j] + nums[k]
#                 if sum == target:
#                     s.add((nums[i], nums[j], nums[k]))
#                     j += 1
#                     k -= 1
#                 elif sum < target:
#                     j += 1
#                 else:
#                     k -= 1
#         output = list(s)
#         return output
```
