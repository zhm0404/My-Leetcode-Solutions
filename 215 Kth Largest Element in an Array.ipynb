Given an integer array nums and an integer k, return the kth largest element in the array.

Note that it is the kth largest element in the sorted order, not the kth distinct element.

Can you solve it without sorting?

```python
# Attempt 1
from typing import List
class Solution2:
    def findKthLargest(self, nums: List[int], k: int) -> int:
        start = 0
        end = len(nums)
        midOrigin = 0
        while True:
            # print(nums[start:end], "k:", k)
            partition = self.partition(nums, start, end, midOrigin)
            # print("After partition:",nums, "Old:",midOrigin, "New:", partition, "k:", k)

            # if the target is at the partition point
            if end - partition == k:
                return nums[partition]
            
            # if the target falls into the right segment
            elif end - partition > k:
                start = partition
                # look for a different partition point
                partitionChanged = False
                for i in range(start,end):
                    if nums[i]!=nums[partition]:
                        midOrigin = i
                        partitionChanged = True
                        break
                # if all the values in nums[start:end] are the same as the value at partition point
                if not partitionChanged: 
                    return nums[partition]
                
            # if the target falls into the left segment
            elif end - partition < k:
                k = k - (end - partition)
                end = partition
                midOrigin = start

                

    # one round of quick sorting
    def partition(self, nums:List[int], start:int, end:int, midOrigin:int) -> int:
        # Swap pivot with the first element if it's not already at the beginning
        if midOrigin != start:
            nums[start], nums[midOrigin] = nums[midOrigin], nums[start]
        low = start
        high = end-1
        pivot = nums[start]

        while low < high:
            while nums[high] >= pivot and low < high:
                high -= 1
            nums[low] = nums[high]
            while nums[low] < pivot and low < high:
                low += 1
            nums[high] = nums[low]
        
        nums[low] = pivot
        return low

# Attempt 2
class Solution1:
    def findKthLargest(self, nums: List[int], k: int) -> int:
        start = 0
        end = len(nums)
        midOrigin = 0
        while True:
            partition1, partition2 = self.partition(nums, start, end, midOrigin)
            
            # if the target falls into the right segment
            if end - partition2 >= k:
                start = partition2
                midOrigin = start
                
            # if the target falls into the left segment
            elif end - partition1 < k:
                k = k - (end - partition1)
                end = partition1
                midOrigin = start

            else:
                return nums[partition1]
                

    # one round of quick sorting
    def partition(self, nums:List[int], start:int, end:int, midOrigin:int) -> tuple:
        pivot = nums[midOrigin]

        low = start
        mid = start
        high = end-1

        while mid <= high:
            if nums[mid] < pivot:
                nums[low], nums[mid] = nums[mid], nums[low]
                low += 1
                mid += 1
            elif nums[mid] == pivot:
                mid += 1
            else:
                nums[mid], nums[high] = nums[high], nums[mid]
                high -= 1
        return low, mid
    
# Attempt 3
import random
class Solution:
    def findKthLargest(self, nums, k):
        if not nums: return
        pivot = random.choice(nums)
        left =  [x for x in nums if x > pivot]
        mid  =  [x for x in nums if x == pivot]
        right = [x for x in nums if x < pivot]
        
        L, M = len(left), len(mid)
        
        if k <= L:
            return self.findKthLargest(left, k)
        elif k > L + M:
            return self.findKthLargest(right, k - L - M)
        else:
            return mid[0]
```
